**k-gram based linear interpolation**

term frequency

```python
import re
from collections import Counter
# Tokenize the cleaned text into words
words = re.findall(r'\w+', article_text.lower())  # Split by words and convert to lowercase

# Create a dictionary of word counts
word_count_dict = Counter(words)
```

```python
word_count_dict["many"],len(word_count_dict)
```

bigram

```python
# Create bi-grams (adjacent pairs of words)
bi_grams = [(words[i], words[i + 1]) for i in range(len(words) - 1)]
# Create a dictionary of bi-grams
bi_gram_count_dict = Counter(bi_grams)
```

```python
bi_gram_count_dict
```

```python
# Filter and print bigrams with count >= 2
i0=0
for bigram, count in bi_gram_count_dict.items():
    if count >= 2:
        print(f"Bigram: {bigram}, Count: {count}")
        i0+=1
print("total relevant bi phrases",i0)
```

```python
bi_gram_count_dict.get(('the', 'public'), 0)
```

```python
len(bi_gram_count_dict), len(word_count_dict)
```

3-gram

```python
# Create bi-grams (adjacent pairs of words)
tri_grams = [(words[i], words[i + 1],words[i+2]) for i in range(len(words) - 2)]
# Create a dictionary of bi-grams
tri_gram_count_dict = Counter(tri_grams)
```

```python
# Filter and print bigrams with count >= 2
i0=0
for trigram, count in tri_gram_count_dict.items():
    if count >= 2:
        print(f"Trigram: {trigram}, Count: {count}")
        i0+=1
print("total relevant tri phrases",i0)
```

Memory space observation: it is not O(N^k)

```python
len(bi_gram_count_dict), len(word_count_dict),len(tri_gram_count_dict)
```

Now we have data prepared for our use.
We need to find 3 parameters a,b,c s.t. c>b>a and a+b+c=1.

```python
from collections import defaultdict
bidict=defaultdict(list)
for w1, w2 in bi_grams:
    bidict[w1].append(w2)
tridict=defaultdict(list)
for w1, w2, w3 in tri_grams:
    tridict[(w1,w2)].append(w3)
```

```python
sorted_count=[(count,word) for word,count in word_count_dict.items()]
sorted_count.sort()
sorted_count.reverse()
```

```python
sorted_count
```

```python
def prediction(s,k):
  s=s.lower()
  ss=s.strip().split()
  m=len(ss)
  l1=bidict.get(ss[-1],[])
  l2=sorted_count[:k]
  #print(l1,l2)
  res=defaultdict(list)
  for e in l1:
    res[e]=bi_gram_count_dict.get((ss[-1],e),0)*0.25
  for count,word in l2:
    #print(word,word_count_dict.get(word,0))
    if res.get(word,0):
      #print(word,"word",res[word])
      res[word]+=word_count_dict.get(word,0)*1e-10
    else:
      res[word]=word_count_dict.get(word,0)*1e-10
  if m>=2:
    w1,w2=ss[-2],ss[-1]
    l3=tridict.get((w1,w2),[])
    for e in l3:
      if res.get(e,0):
        res[e]+=0.75*tri_gram_count_dict.get((w1,w2,e),0)
      else:
        res[e]=0.75*tri_gram_count_dict.get((w1,w2,e),0)

  res=[(count,word) for word,count in res.items()]
  res.sort()
  res.reverse()
  return res[:k]

```

```python
prediction("This is for",5)
```

```python
p=["the more you study", "go there", " I have a", "It ","The people", "When he heard", "What if", "You are very", "do it", "I want to play with"]
```

```python
for line in p:
  print(line,*prediction(line,5))
```
